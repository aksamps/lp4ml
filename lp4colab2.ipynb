#### range() -> iterable , returns numbers

1. range(stop)
2. range(start, stop)
3. range(start, stop, step)

range() function in **Python generates** an immutable sequence of numbers, commonly used for looping a specific number of times in for-loops.

0. start, stop and step can be int only. else value error
1. The sequence does not include the stop value.​
2. If step is negative, the range counts downward.​
3. Useful for generating sequences for iteration, indexing, or list creation.

range() is an essential tool for producing sequences of numbers efficiently in Python.

for i in range(5):
    print(i)  # Output: 0, 1, 2, 3, 4


for i in range(2, 6):
    print(i)  # Output: 2, 3, 4, 5


for i in range(1, 10, 2):
    print(i)  # Output: 1, 3, 5, 7, 9
    
1. The step must be a non-zero integer. (range(1, 10, 0) # ValueError: range() arg 3 must not be zero
)
2. With positive step, the sequence stops before reaching stop.
3. With negative step, the sequence counts down and stops before reaching stop.
4. Using step enables generating sequences with custom increments or decrements.    

````python
for i in range(1, 10, 2):    print(i)  # Output: 1, 3, 5, 7, 9

for i in range(10, 0, -2):   print(i)

````

#### slicing -> getting a range (subsequence) of elements
```
syntex:
sequence[start:stop:step]
start -> index to begin (inclusive) default = 0
stop -> index to end (exclusive) default = end of the list in the direction of traversing
step -> interval(optional, default = 1)
```


````python
data=[1,2,3,4,5,4,6]
d=data[2:4:1]
e=data[::-1]
print(id(d), id(data)) # different list objects
print(e, id(e)) # created a reverse of the list
f=data[4:-len(data):-1] #tried -velen(list) instead of 0
print(f, id(f))

print(data.index(4))  # index function of the list returns the index of the element in the list
xx=data[len(data)-data.index(5)::-1] #7-4=3
print(xx)
````
