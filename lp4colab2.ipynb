range() function in **Python generates** an immutable sequence of numbers, commonly used for looping a specific number of times in for-loops.

1. The sequence does not include the stop value.​
2. If step is negative, the range counts downward.​
3. Useful for generating sequences for iteration, indexing, or list creation.

range() is an essential tool for producing sequences of numbers efficiently in Python.

for i in range(5):
    print(i)  # Output: 0, 1, 2, 3, 4


for i in range(2, 6):
    print(i)  # Output: 2, 3, 4, 5


for i in range(1, 10, 2):
    print(i)  # Output: 1, 3, 5, 7, 9
    
1. The step must be a non-zero integer. (range(1, 10, 0) # ValueError: range() arg 3 must not be zero
)
2. With positive step, the sequence stops before reaching stop.
3. With negative step, the sequence counts down and stops before reaching stop.
4. Using step enables generating sequences with custom increments or decrements.

The del keyword in Python does not perform a "deep delete" that immediately removes the element from memory. Instead, what it does is remove the reference to the object (such as an element in a list or a variable name).
If there are no other references to that object elsewhere in the program, Python's garbage collector will eventually free the memory used by the object. However, if other references exist, the object remains in memory.
For example:

Using del my_list[2] removes the reference to the element at index 2 from the list.

If no other reference to that element exists, it becomes eligible for garbage collection.

But the actual memory release is managed by Python's garbage collector and happens asynchronously.

The del keyword and the remove() method in Python lists are both used to delete elements, but they differ in how they perform deletion:

del keyword
It is a Python keyword, not a method.

Deletes an element or a slice of elements by index.

Can delete single elements, slices, or even the entire list.

Does not return any value.

Raises IndexError if the specified index is out of range.

Example:
```
python
lst = [10, 20, 30, 40, 50]
del lst[1]        # Deletes element at index 1 (20)
print(lst)        # Output: [10, 30, 40, 50]

del lst[1:3]      # Deletes slice from index 1 up to 3 (30,40)
print(lst)        # Output: [10, 50]
Can also delete entire list: del lst
```
remove() method
It is a list method.
Removes the first occurrence of a specified value.
Does not return any value.
Raises ValueError if the value is not found in the list.
```
Example:

python

lst.remove(20)    # Removes the first 20 found
print(lst)        # Output: [10, 30, 20, 40]
```
lst.remove(100)  # Raises ValueError because 100 not in list

Dynamic arrays are data structures that allow elements to be stored in a contiguous block of memory and can dynamically resize when full, unlike static arrays that have fixed size.

### What is a Dynamic Array?
- Similar to a regular array but the size can be modified at runtime.
- When the array is full and a new element is added, the dynamic array:
  1. Allocates a larger block of memory (usually double the current capacity).
  2. Copies existing elements into this new block.
  3. Deletes or ignores the old block.
  4. Adds the new element in the resized array.
- This resizing mechanism allows efficient amortized append operations.

### How Python Implements Dynamic Arrays
- Python’s **list** is implemented as a dynamic array.
- It uses an underlying fixed-size array that grows automatically as elements are added beyond its capacity.
- The capacity typically grows by doubling when resizing is required.
- Internally, Python uses low-level arrays with contiguous memory for performance.

### Sample Python Dynamic Array Implementation (simplified)
```python
import ctypes

class DynamicArray:
    def __init__(self):
        self.n = 0            # Count of actual elements
        self.capacity = 1     # Current array capacity
        self.A = self.make_array(self.capacity)

    def make_array(self, capacity):
        return (capacity * ctypes.py_object)()

    def append(self, item):
        if self.n == self.capacity:
            self._resize(2 * self.capacity)   # Double capacity
        self.A[self.n] = item
        self.n += 1

    def _resize(self, new_cap):
        B = self.make_array(new_cap)
        for i in range(self.n):
            B[i] = self.A[i]
        self.A = B
        self.capacity = new_cap
```

This example exhibits how a new larger array is created and data migrated when capacity is exceeded.

Python lists use this dynamic array logic internally, providing efficient random access and append while abstracting memory management from the user.

The **copy()** method of a Python list creates a **shallow copy** of the list. This means it returns a new list object containing references to the same elements as the original list, but the list itself is independent.

### How copy() works:
- It creates a new list object.
- The new list contains references to the same elements in the original list.
- Modifying the new list (e.g., adding/removing elements) doesn’t affect the original list.
- However, if the list contains **mutable objects** (like nested lists), modifications to those objects in either list will reflect in both, because the elements themselves are not copied.

### Example:

```python
original = [1, 2, 3]
shallow_copy = original.copy()

shallow_copy.append(4)
print(original)       # Output: [1, 2, 3]
print(shallow_copy)   # Output: [1, 2, 3, 4]
```

### Shallow copy with nested list:

```python
original = [1, [2, 3], 4]
shallow_copy = original.copy()
shallow_copy[1].append(5)

print(original)       # Output: [1, [2, 3, 5], 4]
print(shallow_copy)   # Output: [1, [2, 3, 5], 4]
```

### Summary:
- `copy()` creates a shallow copy of the list.
- The outer list is copied, but inner mutable objects are shared.
- To create a deep copy (fully independent), use `copy.deepcopy()` from the `copy` module.

The copy method is useful when you need a new list to modify without affecting the original list's structure but should be used with caution if the list contains nested mutable objects.

````python
import copy

original = [1, [2, 3], 4]
shallow_copy = original.copy()
print(shallow_copy)
deep_copy = copy.deepcopy(original)
print(deep_copy)
print(id(original[1]), id(shallow_copy[1]), id(deep_copy[1]))
````
