```python
print(int(complex(1,2))) #-> cant be converted to int from complex (int() argument must be a string, a bytes-like object or a real number, not 'complex')
```

```python
# Basic usage
s = "Strae"
print("Using lower():", s.lower())     # Output: strae
print("Using casefold():", s.casefold())  # Output: strasse

# Regular string example
text = "PYTHON is FUN!"
print(text.casefold(),"\n\n")  # Output: python is fun!



'''
The casefold() method in Python is used to convert a string into a lowercase
form that is even more aggressive than the standard lower() method.
It is specifically designed for case-insensitive comparisons, especially when
handling Unicode characters (such as German "") that may not be properly
managed by lower()
'''
```
